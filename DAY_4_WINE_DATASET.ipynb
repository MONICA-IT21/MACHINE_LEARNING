```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.metrics._plot.confusion_matrix import confusion_matrix


%matplotlib notebook
```

```python
wine = datasets.load_wine()
```

```python
df = pd.DataFrame(wine["data"], columns=wine["feature_names"])
df["target"] = wine["target"]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
X = df
y = X.pop("target")
```

```python
X.head()
```

```python
X.shape
```

```python
y.unique()
```

```python
X_train, X_test,  y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=55)
```

```python
X_train.shape
```

```python
X_test.shape
```

**TRAIN CLASSIFIER**

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

|**TUNNING SENSITIVITY OF MODEL TO n_neighbors**

```python
k_range = [1, 3, 5, 7, 9]
scores = [0.85, 0.87, 0.88, 0.86, 0.84]

plt.figure()
plt.xlabel("K count")
plt.ylabel("Model Accuracy")
plt.scatter(k_range, scores)
plt.grid()
plt.xticks([0, 5, 10, 15, 20, 30])  # Use square brackets for indexing
plt.show()
```

```python
test_sizes = [0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1]
knn = KNeighborsClassifier(n_neighbors=5)

plt.figure()

for test_size in test_sizes:
    scores = []

    for i in range(1, 1000):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1 - test_size)
        knn.fit(X_train, y_train)
        scores.append(knn.score(X_test, y_test))

    plt.plot(test_size, np.mean(scores))  # Corrected line

plt.xlabel("Training split %")
plt.ylabel("Accuracy")
plt.show()
```
